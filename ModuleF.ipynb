1. Numerical Differentiation Problem Statement

A numerical grid is an evenly spaced set of points over a function’s domain.
The spacing (step size h) is the distance between adjacent grid points.
In numerical differentiation, we use these points to approximate derivatives of functions.

2. Finite Difference Approximating Derivatives

Finite difference methods approximate the derivative f′(x) using nearby function values:

Forward Difference:
f′(x_j) ≈ (f(x_j+1) − f(x_j)) / h

Backward Difference:
f′(x_j) ≈ (f(x_j) − f(x_j−1)) / h

Central Difference:
f′(x_j) ≈ (f(x_j+1) − f(x_j−1)) / (2h)

Below is an example using forward difference for f(x) = sin(x):


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
%matplotlib inline

# Set step size and grid
h = 0.1
x = np.arange(0, 2*np.pi, h)
y = np.sin(x)

# Compute forward difference approximation
forward_diff = np.diff(y) / h
x_diff = x[:-1]  # grid for the finite difference
exact_solution = np.cos(x_diff)  # since derivative of sin(x) is cos(x)

# Plot the finite difference approximation and the exact derivative
plt.figure(figsize=(12, 8))
plt.plot(x_diff, forward_diff, '--', label='Finite Difference Approximation')
plt.plot(x_diff, exact_solution, label='Exact Derivative (cos(x))')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.legend()
plt.show()

# Compute maximum error between the numerical and exact derivatives
max_error = np.max(np.abs(exact_solution - forward_diff))
print("Maximum error:", max_error)
```

3. Approximating Higher Order Derivatives
Higher order derivatives can also be approximated using finite differences.
For example, the second derivative can be approximated by:

f′′(x_j) ≈ (f(x_j+1) − 2f(x_j) + f(x_j−1)) / h²

Below is an example for f(x) = sin(x) (whose exact second derivative is −sin(x)):


```python
# Use the same grid from above
second_deriv_approx = (y[2:] - 2*y[1:-1] + y[:-2]) / h**2
x_second = x[1:-1]  # corresponding grid points
exact_second = -np.sin(x_second)  # second derivative of sin(x)

# Plot the approximated second derivative vs. the exact solution
plt.figure(figsize=(12, 8))
plt.plot(x_second, second_deriv_approx, 'r--', label='Approximate f\'\'(x)')
plt.plot(x_second, exact_second, 'b-', label='Exact f\'\'(x) = -sin(x)')
plt.xlabel('x')
plt.ylabel("f''(x)")
plt.legend()
plt.show()

max_error_second = np.max(np.abs(exact_second - second_deriv_approx))
print("Maximum error in second derivative:", max_error_second)
```

4. Numerical Differentiation with Noise

Sometimes data comes with noise. In this example, we add noise to f(x) = sin(x)
and plot the noisy data alongside the true function.


```python
x_noise = np.arange(0, 2*np.pi, 0.01)
omega = 110
epsilon = 0.1
y_clean = np.sin(x_noise)
y_noise = y_clean + epsilon * np.cos(omega * x_noise)

plt.figure(figsize=(12, 8))
plt.plot(x_noise, y_noise, 'r-', label='sin(x) + noise')
plt.plot(x_noise, y_clean, 'b-', label='sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
