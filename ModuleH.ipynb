1. ODE Initial Value Problem Statement

An Ordinary Differential Equation  is an equation that relates a function y(t) and its derivatives with respect to an independent variable.
An initial value problem specifies the ODE together with the value of the function  at a starting point.

For example, for the ODE:

dy/dt = f(t, y),

with initial condition:

y(t_0) = y_0,

the goal is to find y(t) that satisfies both the differential equation and the initial condition.


2. Reduction of Order

Many numerical methods (like Euler’s method) are designed for first-order ODEs.
To apply these methods to higher-order ODEs, we reduce the order by introducing new variables.

For example, for a second-order ODE:


y′′ = g(t, y, y′),

we can set:

v = y′,

and rewrite the system as:

{
   y′ = v,
   v′ = g(t, y, v),
}


with initial conditions y(t_0) = y_0 and v(t_0) = y′_0.
This converts the second-order ODE into a system of two first-order ODEs, which can then be solved using standard methods.


3. The Euler Method

The Euler Method is a simple numerical method to solve first-order IVPs. Given an ODE

dy/dt = f(t, y),

with initial condition y(t_0) = y_0, the Euler update step is:

y_(n+1) = y_n + h * f(t_n, y_n),

where h is the step size. Below is an exaple that approximates the solution of the ODE

dy/dt = e^(-t),

with the initial condition y(0) = -1. The exact solution is -e^(-t), so we can compare the approximate result with the exact solution.


```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-poster')

# Define the ODE: dy/dt = exp(-t)
f = lambda t, y: np.exp(-t)

# Set step size and time grid
h = 0.01
t = np.arange(0, 1 + h, h)
y0 = -1

# Allocate solution array and set initial condition
y = np.zeros(len(t))
y[0] = y0

# Euler method loop
for i in range(len(t) - 1):
    y[i+1] = y[i] + h * f(t[i], y[i])

# Compute the exact solution: y(t) = -exp(-t)
exact = -np.exp(-t)

# Plot the approximate and exact solutions
plt.figure(figsize=(12, 8))
plt.plot(t, y, 'b--', label='Euler Approximation')
plt.plot(t, exact, 'g', label='Exact Solution (-exp(-t))')
plt.title('Euler Method Approximation vs. Exact Solution for dy/dt = exp(-t)')
plt.xlabel('t')
plt.ylabel('y')
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
```
